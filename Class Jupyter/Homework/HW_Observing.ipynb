# Observing with `astroplan`

## I have given you:

* Three locations
* An observing date
* Some target objects

```python
import matplotlib.pyplot as plt

import numpy as np
import astropy.units as u
from astropy.time import Time
import pytz

from astroplan.plots import plot_airmass
from astroplan import Observer, FixedTarget, time_grid_from_range
```

# Set up the locations

- [Google](https://www.google.com/), [Google Maps](https://www.google.com/maps), and [Google Earth](https://www.google.com/earth/) are good resources to find Latitude, Longitude, and Timezones
- Make sure you have not switched the Lat and Lon!

```python
# Location 1
npe = Observer(latitude = 78.2233456 * u.deg, 
                      longitude = 15.6535408 * u.deg,
                      timezone = 'Europe/Oslo',   # Central European Time
                      name = "North Pole Expedition Museum"
                     )
```

```python
# Location 2
smb = Observer(latitude = 1.2807531 * u.deg, 
                      longitude = 103.8474177 * u.deg,
                      timezone = 'Asia/Singapore',   # Singapore Standard Time
                      name = "Singapore Musical Box Museum"
                     )
```

```python
# Location 3
rakiura = Observer(latitude = -46.8982053 * u.deg, 
                      longitude = 168.1270134 * u.deg,
                      timezone = 'Pacific/Auckland',   # New Zealand Daylight Time
                      name = "Rakiura Museum"
                     )
```

# Determine the UTC time for **Local Noon** at each of the locations

# Observing Date: **Dec 24, 2024**

* You will have to experiment with the `which =` parameter
* Make sure local noon occurs on the observing date

```python
date = Time("2024-12-24")
date.iso
```

```python
# Location 1
utc_npe = npe.noon(date, which = "nearest")
utc_npe.iso
```

```python
# Location 2
utc_smb = smb.noon(date, which = "nearest")
utc_smb.iso
```

```python
# Location 3
utc_rakiura = rakiura.noon(date, which = "nearest")
utc_rakiura.iso
```

# Determine the length of the observing run at each location

* Length of run is defined by **Astronomical Twilight**
* If your length is negative (or > 24 hr), you have to use a different `which =` parameter above

```python
# Location 1
observe_npe = npe.tonight(utc_npe, horizon = -18 * u.deg)
observing_npe = (observe_npe[1] - observe_npe[0]).to(u.h)

print(f"You can have {observing_npe:.1f} of observation time tonight between \n{observe_npe[0].iso} UTC and {observe_npe[1].iso} UTC")
```

```python
# Location 2
observe_smb = smb.tonight(utc_smb, horizon = -18 * u.deg)
observing_smb = (observe_smb[1] - observe_smb[0]).to(u.h)

print(f"You can have {observing_smb:.1f} of observation time tonight between \n{observe_smb[0].iso} UTC and {observe_smb[1].iso} UTC")
```

```python
# Location 3
observe_rakiura = rakiura.tonight(utc_rakiura, horizon = -18 * u.deg)
observing_rakiura = (observe_rakiura[1] - observe_rakiura[0]).to(u.h)

print(f"You can have {observing_rakiura:.1f} of observation time tonight between \n{observe_rakiura[0].iso} UTC and {observe_rakiura[1].iso} UTC")
```

# What can you see at each location?

## For each location, plot airmass vs. time for all of the targets over the observing night

```python
window_npe = time_grid_from_range(observe_npe, time_resolution = 0.1 * u.h)

fig, ax = plt.subplots(
    figsize = (12, 6), 
    constrained_layout = True
)

for object in targets:
    ax = plot_airmass(object, npe, window_npe)

ax.legend(loc=0,shadow=True);
```

```python
window_smb = time_grid_from_range(observe_smb, time_resolution = 0.1 * u.h)

fig, ax = plt.subplots(
    figsize = (12, 6), 
    constrained_layout = True
)

for object in targets:
    ax = plot_airmass(object, smb, window_smb)

ax.legend(loc=0,shadow=True);
```

```python
window_rakiura = time_grid_from_range(observe_rakiura, time_resolution = 0.1 * u.h)

fig, ax = plt.subplots(
    figsize = (12, 6), 
    constrained_layout = True
)

for object in targets:
    ax = plot_airmass(object, rakiura, window_rakiura)

ax.legend(loc=0,shadow=True);
```
