# Image

```python
import numpy as np
import matplotlib.pyplot as plt

import astropy.io.fits as fits
from astropy.wcs import WCS

from astroplan import FixedTarget
from astroplan.plots import plot_finder_image
from astropy import units as u
```

---

# Start with the object you picked for the Astro Coords Homewok.

- Use `FixedTarget` and `plot_finder_image` to get the objects' FITS file.
- Make sure to set the `fov_radius =` large enough to see all of the object (plus a little extra for context)

```python
messier90 = FixedTarget.from_name('Messier 90')
```

```python
fig, ax = plt.subplots(
    figsize = (6, 6), 
    constrained_layout = True
)

ax, hdu = plot_finder_image(
    messier90, 
    fov_radius= 0.2 * u.deg, 
    ax = ax,
    survey = 'DSS2 Blue'
);
```

```python
ms90_header = hdu.header
ms90_data = hdu.data
```

## Get the World Coordinate System (WCS) data from your FITS file.

```python
wcs = WCS(ms90_header) # adds units to RA/DEC in plot ax
```

---
## Create the following 4 plots:

- Make them each a seperate plot (8 in x 8 in).
- All images should have RA and DEC coords on the labels (use WCS)
- Put a title on each plot
  - **Plot 1**: Your object
  - **Plot 2**: Your object where `(values < MEAN set to min)`
  - **Plot 3**: Your object where `(values < (60% of MAX) set to min)`
  - **Plot 4**: Your object where `(values < (90% of MAX) set to min)`
  
## Actually change the values in the images. Don't just use vmin/vmax

```python
max_value = np.max(ms90_data)
cut60 = 0.6 * max_value
cut90 = 0.9 * max_value
```

```python
fig, ax = plt.subplots(
    subplot_kw={'projection': wcs},
    figsize = (8, 8), 
    constrained_layout = True
)

fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Messier 90 Image")
ax.set_xlabel("RA")
ax.set_ylabel("DEC")

ax.imshow(ms90_data, 
          cmap = plt.cm.Spectral,
          origin = "lower" # sets origin to lower left corner
         )
```

```python
fig, ax = plt.subplots(
    subplot_kw={'projection': wcs},
    figsize = (8, 8), 
    constrained_layout = True
)

fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Messier 90 Image")
ax.set_xlabel("RA")
ax.set_ylabel("DEC")

ax.imshow(ms90_data, 
          cmap = plt.cm.magma,
          origin = "lower", # sets origin to lower left corner
          vmin = ms90_data.mean(),
          vmax = max_value
         )
```

```python
fig, ax = plt.subplots(
    subplot_kw={'projection': wcs},
    figsize = (8, 8), 
    constrained_layout = True
)

fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Messier 90 Image")
ax.set_xlabel("RA")
ax.set_ylabel("DEC")

ax.imshow(ms90_data, 
          cmap = plt.cm.viridis,
          origin = "lower", # sets origin to lower left corner
          vmin = cut60,
          vmax = max_value
         )
```

```python
fig, ax = plt.subplots(
    subplot_kw={'projection': wcs},
    figsize = (8, 8), 
    constrained_layout = True
)

fig.set_constrained_layout_pads(wspace = .15)

ax.set_title("Messier 90 Image")
ax.set_xlabel("RA")
ax.set_ylabel("DEC")

ax.imshow(ms90_data, 
          cmap = plt.cm.gray,
          origin = "lower", # sets origin to lower left corner
          vmin = cut90,
          vmax = max_value
         )
```
