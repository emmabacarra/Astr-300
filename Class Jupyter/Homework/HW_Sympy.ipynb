# SymPy

```python
import sympy as sp
import sympy.plotting as spl
import numpy as np
```

```python
x, t = sp.symbols('x t')
sp.init_printing()
```

---

$$ \Large 
f(x) = 3 \sin{\left(\frac{x}{3} \right)} \cos{\left(3 x \right)}
$$

&nbsp;

$$\Large 
{\displaystyle g(x)=\frac{1}{9}x^{3} + \frac{1}{2}x^{2} + \frac{1}{3}x - \frac{2}{7}\pi}
$$

---

## What is the value of:

$$ \Large
\left.{\frac {d}{dx}\, \left( f(x)\cdot g(x) \right)   }\ \right|_{\ x\, =\, 0}
$$

```python
fox = 3 * sp.sin(x/3) * sp.cos(3*x)
gox = (sp.Rational(1,9))*x**3 + (sp.Rational(1, 2))*x**2 + (sp.Rational(1, 3))*x - (sp.Rational(2, 7))*sp.pi

fox, gox
```

```python
sp.diff((fox*gox), x, 1).subs({x: 0})
```

---

## Plot $f(x)$ and $g(x)$ on the same plot over the interval $[-4 \le x \le 2]$

```python
spl.plot(fox, gox, 
        (x, -4, 2),
        title = "My SymPy Plot", 
        xlabel = 'x', 
        ylabel = 'my function',
        size = (12, 6)
       );
```

---

## What are the value(s) for $x$ where $f(x)\ = \ g(x)$ over the interval $[-4 \le x \le 2]$

```python
guesses = np.array([-4, -3, -2, -1, 0, 1, 2])
intersect = np.array([])

for n in guesses:
    result = sp.nsolve(fox - gox, x, n)
    print(result)
    intersect = np.append(intersect, result)

intersect = np.unique(intersect)
intersect
```

```python
f"F(x) and G(x) intersect at x = {intersect}."
```

## Use `Sympy` to solve this equation for `d`

$$ \Large
T = \left( \frac{L(1-A)}{4\sigma \pi d^{2}} \right)^{\frac{1}{4}}
$$

Use the `sp.Rational()` function for the 1/4 power

```python
A, T, sigma, L, pi, d = sp.symbols('A, T, sigma, L, pi, d')
```

```python
funct = sp.Eq(
    ( (L * (1 - A)) /
        (4 * sigma * pi * (d**2))
    )**sp.Rational(1, 4),
    T
)
funct
```

```python
sp.solve(funct, d)
```
