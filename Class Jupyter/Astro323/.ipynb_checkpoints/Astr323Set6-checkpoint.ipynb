```python
import numpy as np
from astropy.table import QTable
import matplotlib.pyplot as plt
```

```python
dict = {
    'Supernova': np.array([
        '1990O', '1990Y', '1991S', '1991U', '1992J', '1992P', '1992ag', '1992aq', '1992bc', '1992bk', '1992bo', '1993B'
    ]),
    
    'Log(cz)': np.array([
        3.958, 4.066, 4.223, 3.992, 4.137, 3.897, 3.891, 4.481, 3.774, 4.240, 3.736, 4.326
    ]),
    
    'm_Max': np.array([
        16.409, 17.3488, 17.8124, 16.3653, 17.2548, 16.2426, 16.2164, 19.0955, 15.4026, 17.7777, 15.4329, 18.4524
    ])
    
}

table = QTable(dict)
table
```

```python
x = table['Log(cz)']
y = table['m_Max']

fit = np.polyfit(x.value, y.value, 1)
polyfit = np.poly1d(fit)
table['Fit'] = polyfit(x.value)

fig1, ax = plt.subplots(
    figsize = (8, 5), 
    constrained_layout = True
)

ax.set_xlabel("Log(cz)")
ax.set_ylabel(r'Apparent Magnitude $m_{Max}$')
ax.set_title("Supernovae Positions")

ax.set_xlim([3.6, 4.6])
ax.set_ylim([14, 20])

ax.plot(x, y,
        color = "m",
        marker = "o",
        linestyle = "None",
        label = "Supernovae",
        markersize = 10);

for i, txt in enumerate(table['Supernova']):
    ax.annotate(txt, (x[i], y[i]), xytext=(10,10), textcoords='offset points')

ax.plot(x, table['Fit'],
        marker = "None",
        linewidth = 4,
        color = (0.3, 0.1, 0.9, 0.4),
        linestyle = '--',
        label = "Fit to Data")

ax.text(4.3, 15.1, f"Slope of Line: {polyfit[1]:.0f}", fontsize = 12)

ax.legend(loc=2);
```

```python
fig1.savefig('Astr323Set6P1a.png', bbox_inches='tight')
```

```python
redshift = {
    'Supernova': np.array([
        '1995K', '1996E', '1996H', '1996I', '1996J', '1996K', '1996U', '1997ce', '1997cj', '1997ck'
    ]),

    'z': np.array([
        0.48, 0.43, 0.62, 0.57, 0.3, 0.38, 0.43, 0.44, 0.5, 0.97
    ]),
    
    'Log(cz)': np.array([
        5.158, 5.111, 5.27, 5.233, 4.954, 5.057, 5.111, 5.121, 5.176, 5.464
    ]),
    
    'm_Max': np.array([
        23.24, 22.78, 23.76, 23.58, 21.74, 22.96, 23.09, 23.01, 23.45, 25.05
    ])
    
}

table2 = QTable(redshift)
table2
```

```python
### Setting Up the Plot ------------------------------------
fig3, ax = plt.subplots(
    figsize = (8, 5), 
    constrained_layout = True
)

ax.set_xlabel("Log(cz)")
ax.set_ylabel(r'Apparent Magnitude $m_{Max}$')
ax.set_title("Supernovae Positions")

ax.set_xlim([3.5, 6])
ax.set_ylim([14, 26])


### Preparing For Low Redshift Supernovae ------------------------------------
# same data as from part 1

x = table['Log(cz)']
y = table['m_Max']

fit = np.polyfit(x.value, y.value, 1)
polyfit = np.poly1d(fit)
table['Fit'] = polyfit(x.value)

ax.plot(x, y, # plotting data
        color = "m",
        marker = "o",
        linestyle = "None",
        label = "Low Redshift Supernovae",
        markersize = 5);

# plotting & extending the polyfit line:
range = np.linspace(3.5, 6, 50)
ax.plot(range, np.poly1d(np.polyfit(x.value, y.value, 1))(range),
        marker = "None",
        linewidth = 3,
        color = (0.3, 0.1, 0.9, 0.4),
        linestyle = '--',
        label = f"Fit to Low Redshifts (Slope = {polyfit[1]:.0f})")


### Preparing for High Redshift Supernovae ------------------------------------
# new additional data introduced

x2 = table2['Log(cz)']
y2 = table2['m_Max']

fit2 = np.polyfit(x2.value, y2.value, 1)
polyfit2 = np.poly1d(fit2)
table2['Fit'] = polyfit2(x2.value)

ax.plot(x2, y2, # plotting data
        color = "darkorange",
        marker = "o",
        linestyle = "None",
        label = "High Redshift Supernovae",
        markersize = 5);

ax.plot(x2, table2['Fit'], # plotting polyfit
        marker = "None",
        linewidth = 3,
        color = ("dodgerblue", 0.4),
        linestyle = '--',
        label = f"Fit to High Redshifts (Slope = {polyfit2[1]:.0f})")


### Setting Legend ------------------------------------

ax.legend(loc=2);
```

```python
fig3.savefig('Astr323Set6P2b.png', bbox_inches='tight')
```
